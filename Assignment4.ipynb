```python
### MICAH JOHNSON

from google.colab import drive
from pathlib import Path
import json
import pandas as pd
import re
drive.mount("/content/drive")
drivepath = Path() / '/content' / 'drive' / 'MyDrive'

FOOD_REF = json.load(open(drivepath / 'nutrients.json'))
```

```python
food = pd.DataFrame.from_dict(FOOD_REF)[['id','group','nutrients']] # Turn loaded JSON into df
food = food.explode('nutrients') # generate new rows for each nutrient in every food
nutr = food['nutrients'] # Get all nutrients for each food

# turn nutr descriptions into new col
# turn nutr values into new col
# turn nutr units into new col
food = food.assign(description=nutr.apply(lambda x: x['description']), \
                   value=nutr.apply(lambda x: x['value']), \
                   units=nutr.apply(lambda x: x['units'])).\
                   drop(columns=['nutrients']).drop_duplicates() # drop old nutr col
food
```

```python
import plotly.express as pltx
```

```python
protein = food.loc[food['description']=='Protein'] # get all rows in which the description fields are protien
protein = protein.groupby('group').sum().reset_index() # sum up all the protein in every group of food
# Plot the sums by group
pltx.bar(protein, x=protein['group'], y=protein['value'], \
         title="Protein in Various Food Groups", \
         labels = {'value':'Protein (grams)', 'group':'Food Group'})
```

```python
maxNutrs = food.sort_values('value', ascending=False).groupby\
         ('description', as_index=False).first()[['group','description','value','units']] # sort by values then group by description to find the highest nutrient value of each food group
maxNutrs
```
